```python
import numpy as np
import pandas as pd
import seaborn as snr
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/tips.csv")
```

```python
data.head()
```

```python
snr.pairplot(data,x_vars=['total_bill','smoker','sex'],y_vars=['tip'],size=5)
```

```python
#violin plot
snr.catplot(x='total_bill',y='day',hue='sex',kind='violin',data=data)
```
