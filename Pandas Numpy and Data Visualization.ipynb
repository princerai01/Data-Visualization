```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as snr

iris=snr.load_dataset("iris")
iris.head()

```

```python
#line graph
x=iris.plot(figsize=(15,8),title="Visulisation on iris dataset")
x.set_xlabel('x_axis')
x.set_ylabel('y_axis')

```

```python
data=iris.drop(['species'],axis=1)
data.head()
```

```python
#bar graph
data.plot(kind='bar')
```

```python
#bar graph for the 0th row
data.iloc[0].plot(kind='bar')
```

```python
#bar graph for the 4th row
data.iloc[0:4].plot(kind='bar')
```

```python
#histogram
data.diff().plot(kind='hist',bins=200)
```

```python
data['petal_length'].diff().plot(kind='hist',bins=200)
```

```python
data['petal_length'].diff().plot(kind='hist',bins=5)
```

```python
data['sepal_length'].diff().plot(kind='hist',bins=200)
```

```python
#pie Chart
data.plot.pie(subplots=True,figsize=(10,10))
```

```python
data.iloc[0].plot.pie(figsize=(10,10),fontsize=15,autopct='%.2f')
```

```python
data.iloc[0:2].plot.pie(subplots=True,figsize=(10,10),fontsize=15,autopct='%.4f')
```

```python
data.iloc[0:3].plot.pie(subplots=True,figsize=(10,10),fontsize=10,autopct='%.4f')
```

```python
#Scatter plot
data.plot.scatter(x='sepal_length',y='petal_length')
```

```python
data.plot.scatter(x='petal_length',y='petal_width',color='r')
```

```python
data.plot.scatter(x='sepal_length',y='sepal_width',color='g')
```

```python
data.plot.scatter(x='sepal_length',y='petal_width',c='sepal_width')
```

```python
#relational plot
my_data=pd.read_csv("/content/tips.csv")
```

```python
my_data.head()
```

```python
#relational plot
snr.relplot(x='total_bill',y='tip',hue='smoker',data=my_data)
```

```python
snr.relplot(x='total_bill',y='tip',hue='smoker',style='sex',data=my_data)
```

```python
snr.relplot(x='total_bill',y='tip',hue='smoker',style='time',size='size',data=my_data)
```

```python
snr.relplot(x='total_bill',y='tip',hue='smoker',style='sex',col='time',data=my_data)
```

```python
snr.relplot(x='total_bill',y='tip',hue='smoker',style='sex',col='day',data=my_data)
```

```python
#catplot
snr.catplot(x='tip', y='time', data=my_data)
```

```python
snr.catplot(x='tip', y='sex', data=my_data)
```

```python
snr.catplot(x='total_bill', y='day', data=my_data)
```

```python
#box plot
snr.catplot(x='day', y='tip',kind='box',hue='sex',data=my_data)
```
